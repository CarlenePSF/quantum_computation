```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes Físicas
h = 6.62607015e-34  # Constante de Planck (Joules * segundo)
c = 3.0e8           # Velocidade da luz no vácuo (m/s)
eV = 1.60218e-19    # Conversão de Joules para eV (1 eV = 1.60218 x 10^-19 J)

def energia_foton(frequencia):
    """Calcula a energia de um fóton dada sua frequência."""
    return h * frequencia

def comprimento_de_onda_frequencia(comprimento_onda):
    """Converte o comprimento de onda em metros para a frequência."""
    return c / comprimento_onda

def efeito_fotoeletrico(comprimento_onda, funcao_trabalho):
    """
    Simula o efeito fotoelétrico.
    
    Parameters:
    comprimento_onda (float): Comprimento de onda da luz (em metros).
    funcao_trabalho (float): Função trabalho do material (em eV).
    
    Returns:
    str: Resultado do efeito fotoelétrico.
    """
    # Converter função trabalho de eV para Joules
    funcao_trabalho_joules = funcao_trabalho * eV
    
    # Calcular a frequência da luz
    frequencia = comprimento_de_onda_frequencia(comprimento_onda)
    
    # Calcular a energia do fóton
    energia_foton_joules = energia_foton(frequencia)
    
    # Verificar se a energia do fóton é suficiente para liberar elétrons
    if energia_foton_joules > funcao_trabalho_joules:
        energia_cinetica = energia_foton_joules - funcao_trabalho_joules
        return f"Efeito fotoelétrico ocorre! Energia cinética do elétron: {energia_cinetica / eV:.2f} eV"
    else:
        return "Fótons não têm energia suficiente para liberar elétrons."

def plot_energia_frequencia():
    """Plota a energia do fóton em função da frequência da luz."""
    # Geração de valores de frequência (de 1e14 a 1e15 Hz)
    frequencias = np.linspace(1e14, 1e15, 500)
    
    # Cálculo das energias correspondentes
    energias = energia_foton(frequencias) / eV  # Converter para eV
    
    # Plot do gráfico
    plt.figure(figsize=(8, 6))
    plt.plot(frequencias, energias, label="Energia do fóton (em eV)", color="b")
    plt.title("Energia do Fóton em Função da Frequência")
    plt.xlabel("Frequência (Hz)")
    plt.ylabel("Energia (eV)")
    plt.grid(True)
    plt.legend()
    plt.show()


```

```python
# Exemplo de uso:
comprimento_onda = 500e-9  # 500 nm (luz visível - verde)
funcao_trabalho = 2.3      # Função trabalho do material em eV (ex: para o sódio)

resultado = efeito_fotoeletrico(comprimento_onda, funcao_trabalho)
print(resultado)

# Plotar o gráfico
plot_energia_frequencia()
```
